###### Define algorithmic bias and provide two examples of how it manifests in AI systems.

Algorithmic bias refers to systematic and unfair distortions in AI outputs that arise when models reflect, amplify, or embed inequities present in their training data, design choices, or deployment context.

*Example 1*
Case: In 2018, Amazon discontinued its AI recruiting tool after discovering it systematically downgraded resumes that included terms associated with women, such as “women’s chess club captain.”

Origin of bias:
The model was trained on resumes submitted over a 10-year period, most of which came from men (reflecting historical gender imbalances in tech roles). As a result, the AI learned to prefer male-associated language patterns and penalize female-coded terms.

Impact:
Qualified female candidates were ranked lower or excluded, perpetuating gender disparities in technical recruitment. This not only risked discriminatory hiring practices but also reinforced historical underrepresentation in engineering teams.

*Example 2*
Case: A 2019 study published in Science revealed that a widely used healthcare risk algorithm in the US underestimated the health needs of Black patients.

Origin of bias:
The model used healthcare costs as a proxy for health needs. However, due to systemic disparities, Black patients historically incur lower healthcare spending despite equal or greater health needs, leading the model to falsely infer they were healthier.

Impact:
Black patients were less likely to be identified for high-risk care management programs, perpetuating unequal access to proactive care and exacerbating health inequities at scale.

###### Explain the difference between transparency and explainability in AI. Why are both important?

Transparency is about providing clarity into how an AI system is built and operates, including its data sources, design choices, and processes, while
Explainability is about providing clarity into why an AI system produces a specific output or decision, making its predictions understandable and interpretable to humans.

Both transparency and explainability are crucial and complementary because transparency without explainability means that while stakeholders may understand how a model was built and its technical pipeline, they still cannot discern why it makes specific decisions, limiting trust, usability, and recourse in practice. Conversely, explainability without transparency allows stakeholders to see why a particular output was produced but leaves them unaware of the underlying data sources, design assumptions, or training processes that shape systemic behavior, making it impossible to identify, audit, or address broader risks and biases. Together, they ensure AI systems are not only technically accountable and ethically sound but also operationally trustworthy and justifiable in real-world deployment.

###### How does GDPR (General Data Protection Regulation) impact AI development in the EU?

GDPR impacts AI by requiring data minimization, restricting solely automated decisions that significantly affect individuals, ensuring explainability of AI outputs, and enforcing accountability through privacy by design. This means AI teams must justify data use, integrate human oversight in impactful decisions, provide meaningful explanations of model outputs to users, and embed privacy safeguards and bias audits throughout development to ensure compliance, user trust, and ethical deployment within the EU.

## Part 2: Case Study Analysis

###### Case 1: Biased Hiring Tool

The bias in Amazon’s hiring tool stemmed from multiple interlinked causes beyond just historical data imbalance. The root issue was training data representation bias, as the model learned from resumes submitted over a decade, predominantly by men, embedding male-dominated language patterns and career trajectories as proxies for success. Proxy feature leakage further compounded this problem, where female-coded terms such as “women’s chess club captain” became negative signals, as these terms deviated from patterns associated with past hires. Additionally, there was a model objective misalignment, since it optimized for replicating previous hiring decisions rather than identifying the most qualified candidates irrespective of gender, thus reinforcing systemic bias. Finally, had it been deployed, it risked creating a feedback loop, further entrenching gender disparities in future data and decisions.

To address these issues, three technically feasible fixes can be proposed. A data-centric approach would involve counterfactual data augmentation by introducing synthetic or real examples that reverse gender-coded terms or ensure balanced representation of female candidates with similar qualifications, though care is needed to maintain realism and avoid noise. A model-centric approach would apply fairness-constrained learning or adversarial debiasing, training the model to penalize gender-correlated representations; however, this requires careful tuning to avoid unacceptable trade-offs in predictive performance. Lastly, an evaluation-centric approach would implement post-processing fairness corrections, such as adjusting model outputs to equalize selection rates across genders through methods like Reject Option Classification, though this only mitigates output disparities without correcting internal model bias.

To evaluate these fixes effectively, concrete fairness metrics should include demographic parity to check for equal selection rates between genders, equal opportunity to ensure equally qualified male and female candidates have similar chances of selection, and disparate impact ratio to assess the proportional fairness of outcomes in line with regulatory thresholds. Together, these targeted strategies and metrics ensure AI hiring systems are not only technically sound but also ethically robust and compliant with fairness mandates in real-world deployment.

###### Facial Recognition in Policing

Ethical Risks
Facial recognition in policing carries multifaceted ethical risks beyond wrongful arrests. Firstly, higher misidentification rates for minorities, especially Black and darker-skinned individuals, risk unjustified stops, detentions, and criminal records, as highlighted in studies such as the Gender Shades audit showing drastically lower accuracy for darker female faces. Secondly, disproportionate surveillance arises as these systems are often deployed in already over-policed communities, amplifying historical inequities and embedding systemic biases into technological infrastructures. Thirdly, chilling effects on civil liberties occur when pervasive facial surveillance deters individuals from exercising freedoms of assembly, protest, or religious gathering, knowing their identities are constantly tracked. For example, in Hong Kong, protesters used masks to avoid facial recognition tracking, underscoring fears of political retribution. Finally, reinforcement of systemic discrimination occurs as these biased datasets and deployment patterns create a self-fulfilling cycle: more surveillance leads to more recorded offences in certain demographics, feeding back into predictive policing tools and justifying further targeted surveillance, thus exacerbating social injustice.

Responsible Deployment Policies
To deploy facial recognition systems responsibly in policing, robust, multi-layered policies are essential. Technically, implement strict accuracy and bias audits disaggregated by race, gender, and age before any operational use, with thresholds for deployment halting if disparities exceed acceptable levels. Governance-wise, establish independent oversight bodies with legal authority to review deployment decisions, assess proportionality, and halt use if ethical standards are violated. Transparency mechanisms should include clear public disclosures about where, when, and why facial recognition is used, ensuring affected communities are informed. Public engagement requires inclusive consultations, especially with marginalized groups disproportionately impacted, to assess societal acceptability and integrate community perspectives into deployment decisions. To ensure accountability and redress, law enforcement agencies must implement clear protocols enabling individuals to contest wrongful identifications, including immediate notification upon a facial recognition match, accessible appeal channels, and mandatory human verification before action is taken. Additionally, policies should define strict use-case limitations, banning real-time surveillance in public spaces unless under court-approved, narrowly defined, and time-bound circumstances. Together, these measures ensure that facial recognition, if deployed, aligns with democratic values, respects civil liberties, and mitigates technological amplification of systemic biases.

```python
import os
import requests
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from aif360.datasets import CompasDataset
from aif360.metrics import ClassificationMetric, BinaryLabelDatasetMetric
from aif360.algorithms.preprocessing import Reweighing
from aif360.algorithms.postprocessing import RejectOptionClassification
from aif360.algorithms.inprocessing import AdversarialDebiasing
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()

# ---------------------------------------------
# Function: Download COMPAS dataset
# ---------------------------------------------
def download_compas_data(url, destination_path):
    os.makedirs(os.path.dirname(destination_path), exist_ok=True)
    response = requests.get(url)
    if response.status_code == 200:
        with open(destination_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded {url} to {destination_path}")
    else:
        print(f"Failed to download {url}. Status code: {response.status_code}")

# ---------------------------------------------
# Function: Load dataset using CompasDataset
# ---------------------------------------------
def load_compas_dataset(csv_path):
    # Using the dedicated CompasDataset class which handles data loading and cleaning
    dataset = CompasDataset()
    return dataset

# ---------------------------------------------
# Function: Apply Reweighing pre-processing
# ---------------------------------------------
def reweigh_dataset(dataset, privileged_groups, unprivileged_groups):
    RW = Reweighing(unprivileged_groups=unprivileged_groups,
                    privileged_groups=privileged_groups)
    return RW.fit_transform(dataset)

# ---------------------------------------------
# Function: Train Logistic Regression model
# ---------------------------------------------
def train_logreg(X, y):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    clf = LogisticRegression(solver='liblinear')
    clf.fit(X_scaled, y)
    return clf, scaler

# ---------------------------------------------
# Function: Evaluate fairness and accuracy
# ---------------------------------------------
def evaluate_fairness(original_dataset, predicted_dataset, privileged_groups, unprivileged_groups, stage_desc):
    print(f"\n=== Fairness and Accuracy Metrics: {stage_desc} ===")
    metric = ClassificationMetric(original_dataset, predicted_dataset,
                                  unprivileged_groups=unprivileged_groups,
                                  privileged_groups=privileged_groups)
    print("Accuracy: {:.3f}".format(metric.accuracy()))
    print("Disparate Impact Ratio: {:.3f}".format(metric.disparate_impact()))
    print("Equal Opportunity Difference: {:.3f}".format(metric.equal_opportunity_difference()))

# ---------------------------------------------
# Function: Apply Reject Option Classification
# ---------------------------------------------
def reject_option_postprocessing(original_dataset, pred_dataset, privileged_groups, unprivileged_groups):
    ROC = RejectOptionClassification(unprivileged_groups=unprivileged_groups,
                                     privileged_groups=privileged_groups,
                                     low_class_thresh=0.01, high_class_thresh=0.99,
                                     num_class_thresh=100, num_ROC_margin=50,
                                     metric_name="Statistical parity difference",
                                     metric_ub=0.05, metric_lb=-0.05)
    ROC = ROC.fit(original_dataset, pred_dataset)
    return ROC.predict(pred_dataset)

# ---------------------------------------------
# Function: Apply Adversarial Debiasing
# ---------------------------------------------
def adversarial_debiasing_train(dataset_transf, privileged_groups, unprivileged_groups):
    tf.reset_default_graph() # Reset the default graph
    sess = tf.Session()
    adv_debiasing = AdversarialDebiasing(privileged_groups=privileged_groups,
                                         unprivileged_groups=unprivileged_groups,
                                         scope_name='adv_debiasing',
                                         debias=True,
                                         sess=sess)
    adv_debiasing.fit(dataset_transf)
    pred_adv = adv_debiasing.predict(dataset_transf)
    sess.close()
    return pred_adv

# ============================
# Main Pipeline Execution
# ============================

if __name__ == "__main__":
    # Define constants
    url = "https://raw.githubusercontent.com/propublica/compas-analysis/master/compas-scores-two-years.csv"
    destination_path = "./compas-scores-two-years.csv"

    # Download dataset
    download_compas_data(url, destination_path)

    # Load dataset
    dataset = load_compas_dataset(destination_path)

    # Define protected groups
    # Assuming 'race' is the protected attribute and 0 represents the privileged group (Caucasian)
    # and 1 represents the unprivileged group (African-American) based on common usage with this dataset
    privileged_groups = [{'race': 0}]
    unprivileged_groups = [{'race': 1}]


    # Pre-processing: Reweighing
    dataset_transf = reweigh_dataset(dataset, privileged_groups, unprivileged_groups)

    # Model training: Logistic Regression
    X = dataset_transf.features
    y = dataset_transf.labels.ravel()
    clf, scaler = train_logreg(X, y)

    # Predictions
    X_scaled = scaler.transform(X)
    y_pred = clf.predict(X_scaled)
    y_proba = clf.predict_proba(X_scaled) # Get prediction probabilities

    # Create predicted dataset
    # Ensure the predicted dataset has the same features and protected attributes as the transformed dataset
    pred_dataset = dataset_transf.copy(deepcopy=True)
    pred_dataset.labels = y_pred.reshape(-1, 1)
    # Explicitly set features and protected attributes to match dataset_transf
    pred_dataset.features = dataset_transf.features
    pred_dataset.protected_attributes = dataset_transf.protected_attributes
    # Add prediction probabilities to the predicted dataset
    fav_label_idx = int(dataset_transf.favorable_label)
    pred_dataset.scores = y_proba[:, fav_label_idx].reshape(-1, 1)


    # Evaluate pre-post mitigation fairness
    evaluate_fairness(dataset_transf, pred_dataset, privileged_groups, unprivileged_groups,
                      "After Reweighing + Logistic Regression")

    # Post-processing: Reject Option Classification
    # Ensure pred_dataset_for_roc has the same features and protected attributes as dataset_transf
    pred_dataset_for_roc = dataset_transf.copy(deepcopy=True)
    pred_dataset_for_roc.labels = y_pred.reshape(-1, 1)
    pred_dataset_for_roc.features = dataset_transf.features
    pred_dataset_for_roc.protected_attributes = dataset_transf.protected_attributes
    # Add prediction probabilities to the predicted dataset for ROC
    fav_label_value = dataset_transf.favorable_label

    # Get index of favorable label within clf.classes_
    fav_label_index = np.where(clf.classes_ == fav_label_value)[0][0]

    # Assign scores using integer index
    pred_dataset_for_roc.scores = y_proba[:, fav_label_index].reshape(-1, 1)


    pred_dataset_post = reject_option_postprocessing(dataset_transf, pred_dataset_for_roc,
                                                     privileged_groups, unprivileged_groups)

    # Evaluate post-processing results
    evaluate_fairness(dataset_transf, pred_dataset_post, privileged_groups, unprivileged_groups,
                      "After Reject Option Classification")

    # In-processing: Adversarial Debiasing
    # Ensure dataset_transf and pred_dataset for adversarial debiasing have aligned features and protected attributes
    dataset_transf_for_adv = dataset_transf.copy(deepcopy=True)
    pred_dataset_for_adv = dataset_transf.copy(deepcopy=True)
    pred_dataset_for_adv.labels = y_pred.reshape(-1, 1)
    pred_dataset_for_adv.features = dataset_transf.features
    pred_dataset_for_adv.protected_attributes = dataset_transf.protected_attributes


    pred_adv = adversarial_debiasing_train(dataset_transf_for_adv, privileged_groups, unprivileged_groups)

    # Evaluate adversarial debiasing results
    evaluate_fairness(dataset_transf, pred_adv, privileged_groups, unprivileged_groups,
                      "After Adversarial Debiasing")

# ============================
# End of Pipeline
# ============================
```

### AI Fairness Audit Report: COMPAS Recidivism Risk Score Bias Analysis

Our audit of the COMPAS recidivism risk score dataset revealed significant racial bias, particularly manifesting as disproportionately higher false positive rates for African-American (unprivileged) defendants compared to Caucasian (privileged) defendants. Specifically, African-American individuals were more likely to be incorrectly predicted as high risk despite not reoffending. This form of bias reflects algorithmic unfairness where incorrect positive classifications (false positives) disproportionately affect one group, raising critical concerns regarding equitable treatment.

In a criminal justice context, such disparities translate to concrete and evere societal harms. A false positive in this setting implies that an individual is eemed to pose a higher risk to public safety than reality indicates, potentially resulting in harsher bail conditions, longer pretrial detention, or denial of parole. Beyond individual harm, this undermines public trust in AI-assisted decision-making, perpetuates systemic racial disparities within judicial processes, and may contribute to broader structural injustices, eroding confidence in legal institutions meant to protect civil rights equitably.

To remediate these biases, two categories of interventions are proposed. First,at the data level, implement reweighing or sampling techniques to adjust for
historical imbalances and ensure representation across groups reflects real-world base rates without embedding discriminatory patterns. This may include fairness-aware pre-processing such as reweighing to balance group contributions in model training. Second, at the model level, integrate fairness-constrained optimization during model training to enforce parity in error rates or deploy in-processing algorithms like adversarial debiasing to mitigate learned discriminatory patterns while maintaining predictive utility. Furthermore, post-processing methods, such as Reject Option Classification, can adjust decision thresholds to reduce disparate impacts in final outputs.

Overall, systematically integrating such fairness interventions, alongside continuous bias monitoring, is crucial to ensure AI tools within criminal justice contexts promote equitable, just, and socially responsible outcomes.
